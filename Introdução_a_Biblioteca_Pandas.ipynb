```
import pandas as pd
```

```
#df = dataframe
df = pd.read_csv("/content/drive/MyDrive/Datasets/Gapminder.csv", error_bad_lines=False, sep=";")
```

```
#Visualizando as 5 primeiras linhas
df.head()
```

```
df = df.rename(columns={"country":"País", "continent": "Continente", "year":"Ano", "lifeExp":"Expectativa de Vida", "pop": "População", "gdpPercap":"PIB"})
```

```
df.head(10)
```

```
#Total de linhas e de colunas
df.shape
```

```
#Saber quais são as colunas
df.columns
```

```
#Saber o tipo de dados (obs: object = String)
df.dtypes
```

```
#Para mostrar as últimas 5 linhas (se não especificar o número de últimas linhas dentro do parenteses. Ex: se botar 10, apareceram as 10 últimas linhas)
df.tail(10)
```

```
df.describe()
```

```
#Retornar os valores unicos na coluna especificada
df["Continente"].unique()
```

```
#Retornar valores só de um continente. Cria uma variável primeiro
Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head()
```

```
Oceania["Continente"].unique()
```

```
#Agrupamento de dados. Quantos países em cada continente
df.groupby("Continente")["Pais"].nunique()
```

```
#Qual a expectativa de vida media para cada ano
df.groupby("Ano")["Expectativa de Vida"].mean()
```

```
#Como saber a média de uma coluna.. Especifica a coluna e acrescenta .mean()
df["PIB"].mean()
```

```
#Como saber a soma de uma coluna. Especifica a coluna e acrescenta .sum()
df ["PIB"].sum()
```
