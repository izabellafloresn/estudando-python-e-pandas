<strong> Planilhas do Excel em Pandas </strong>


```python
#Importando a biblioteca Pandas
import pandas as pd
```

```python
#Leitura dos arquivos que foi feito o Upload
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel ("Recife.xlsx")
df5 = pd.read_excel ("Salvador.xlsx")
```

```python
df1.head()
```

```python
#Juntando todos os arquivos importados (concat)
df = pd.concat([df1, df2,df3, df4, df5])
```

```python
#Exibindo as 5 primeiras linhas 
df.head()
```

```python
#Para ver as 5 ultimas linhas
df.tail()
```

```python
#Pegando uma amostra qualquer da planilha
df.sample(5)
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Alterando o tipo de dado, nesse caso passando o LojaID de número inteiro para uma String
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Verificando de há valores nulos ou faltantes
df.isnull().sum()
```

```python
#Verificando a media das vendas 
df["Vendas"].mean()
```

```python
#Criando uma coluna de receitas
df["Receita"] = df ["Vendas"].mul(df["Qtde"])
df ["Receita/Vendas"] = df["Receita"] / df ["Vendas"]
```

```python
df.head()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando as3 maiores receitas
df.nlargest(3, "Receita")
```

```python
#Retornando as 3 menores receitas
df.nsmallest(3, "Receita")
```

```python
#Agrupando receita por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando conjunto de dados (por receitas, nesse caso), do maior para o menor
df.sort_values("Receita", ascending=False).head(10)
```
